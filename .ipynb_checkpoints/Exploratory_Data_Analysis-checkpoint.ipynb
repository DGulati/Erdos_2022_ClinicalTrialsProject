### EFraction Analysis

```python
#Statistical 
data['Comp_EF_0.5'] = (data['Status'] == 1) & (data['EFraction'] <= 0.5)
data['Term_EF_0.5'] = (data['Status'] == 0) & (data['EFraction'] <= 0.5)

Terminated_EF_Percent = np.round(data['Term_EF_0.5'].value_counts()[True]/(data['Term_EF_0.5'].value_counts()[True]+data['Comp_EF_0.5'].value_counts()[True])*100)
Terminated_EF_Percent
```

The total percentage of trials terminated with an EFraction less than or equal to 0.5 was 52%.  This means if a clinical trial has less than half the number of participants enrolled than what was expected, then the trial has a 52% chance to fail. 

This leads us to suspect the EFraction is a major contributer to whether a clinical trial will fail.  Thus, in our modeling, we hypothesize an important feature to determine the status of a clinical trial will be the EFraction.

```python
pairplot(data=data, 
             hue='Status',
             y_vars = ['EFraction'],
             x_vars = ['Disease'])
plt.ylim([0,0.5])
plt.title('EFraction 0 to 0.5 vs. Disease')
plt.show()
```

From the plot above, we can see the terminated trials in blue are distributed amongst all three diseases when the EFraction is less than 0.5.

```python
pairplot(data=data, 
             hue='Status',
             y_vars = ['EFraction'],
             x_vars = ['Disease'])
plt.ylim([0.5,1])
plt.title('EFraction 0.5 to 1 vs. Disease')
plt.show()
```

```python
pairplot(data=data, 
             hue='Status',
             y_vars = ['EFraction'],
             x_vars = ['Disease'])
plt.ylim([1,2])
plt.title('EFraction 1 to 2 vs. Disease')
plt.show()
```

From the two plots above, we can see the majority of the clincal trials are completed in orange among all three disease when the EFraction is between 0.5 and 2. 

```python
plt.scatter(data['EFraction'], data['Status'])
plt.xlabel('EFraction')
plt.ylabel('Status')
plt.show()
```

From the plot above, we can see where a logistic regression fit would be appropriate based on the EFraction of a trial to determine the completion status.
