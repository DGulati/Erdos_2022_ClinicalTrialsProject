```python
def giveMeXml(a):
    #print(a)
    contents=os.listdir(str(a))
    result=[]
    for i in contents:
        if os.path.isdir(os.path.join(a, i)):
            for j in giveMeXml(os.path.join(a, i)):
                result.append(j)
        else :
            if i[-4:]=='.xml':
                result.append(os.path.join(a, i))
    return result
```
